### Different ways to Plot

#### Method 1

```python
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x, y)
plt.show()
```

#### Method 2

```python
fig = plt.figure()
ax = fig.add_axes([0.1, 0.1, 0.8, 0.8])
ax.plot(x, y)
plt.show()
```

#### Method 3

```python
fig, ax = plt.subplots()
ax.plot(x, y);
```

## Sample Matplotlib Workflow

```python
# 0. Import matplotlib and get it ready for plotting
%matplotlib inline
import matplotlib.pyplot as plt

# 1. Prepare Data
x = [ 1, 2, 3, 4]
y = [ 11, 33, 22, 44]

# 2. Set up Plot
fig, ax = plt.subplots(figsize=(8, 8))

# 3. Plot Data
ax.plot(x, y)

# 4. Customize Plot
ax.set(title = "Sample Plot",
      xlabel = "x-axis",
      ylabel = "y-axis")

# 5. Sharing
fig.savefig("sample.png")
```

## Subplots

```python
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows = 2, ncols = 2, figsize=(10, 5));
ax1.plot(x, np.sin(x));
ax2.scatter(np.random.random(100), np.random.random(100));
ax3.bar(candies_and_prices.keys(), candies_and_prices.values());
ax4.hist(bell);
```

```python
fig, ax = plt.subplots(nrows = 2, ncols = 2, figsize = (20, 5));
ax[0, 0].hist(bell);
ax[0, 1].plot(x, np.cos(x));
ax[1, 0].scatter(np.random.random(25), np.random.random(25));
ax[1, 1].barh(list(candies_and_prices.keys()), list(candies_and_prices.values()));
```

```python
heart_disease = pd.read_csv("heart-disease.csv")
heart_disease.head()
```

```python
heart_disease.plot.hist(subplots=True, figsize=(10, 50));
```

```python
over_50 = heart_disease[heart_disease['age'] > 50]
over_50.head()
```

```python
over_50.plot(kind= 'scatter', x='age', y='chol', c='target')
```

```python
fig, ax = plt.subplots(figsize=(10, 5))
over_50.plot(kind='scatter', x='age', y='chol', c='target', ax= ax)
ax.set_xlim([50, 80])
```

## Object Oriented Plotting

```python
fig, ax = plt.subplots(figsize=(10, 6))
scatter = ax.scatter(x = over_50['age'], y = over_50['chol'], c = over_50['target'])
ax.set(title = 'Heart Disease and Cholesterol Level', xlabel = 'Age', ylabel = 'Cholesterol')

ax.legend(*scatter.legend_elements(), title='Target')
ax.axhline(over_50['chol'].mean(), linestyle='--');
```

```python
over_50.head()
```

```python
fig, ax = plt.subplots(figsize=(10, 6))
chol_v_lach = ax.scatter(x=over_50['chol'], y=over_50['thalach'], c=over_50['target'])
ax.set(title='Cholesterol and Heart Rate', xlabel = 'Cholesterol', ylabel='Heart Rate')
ax.legend(*chol_v_lach.legend_elements(), title='Target')
ax.axhline(over_50['thalach'].mean(), linestyle='--')
```

```python
over_50.head()
```

```python
fig, (ax0, ax1) = plt.subplots(figsize=(10, 10), ncols=1, nrows=2, sharex=True)

# First Axes
scatter = ax0.scatter(x=over_50['age'], y=over_50['chol'], c=over_50['target'])
ax0.set(title='Heart Disease and Cholesterol Levels', ylabel='Cholesterol')
ax0.legend(*scatter.legend_elements(), title='Target')
ax0.axhline(over_50['chol'].mean(), linestyle='--')

# Second Axes
scatter = ax1.scatter(x=over_50['age'], y=over_50['thalach'], c=over_50['target'])
ax1.set(title='Heart Disease and Max Heart Rate', xlabel='Age', ylabel='Heart Rate')
ax1.legend(*scatter.legend_elements(), title='Target')
ax1.axhline(over_50['thalach'].mean(), linestyle='--');

# Figure title
fig.suptitle("Heart Disease Analysis", fontsize=20, fontweight='bold')
```

```python
plt.style.use('seaborn-v0_8-whitegrid')
fig, (ax0, ax1) = plt.subplots(figsize=(10, 10), ncols=1, nrows=2, sharex=True)

# First Axes
scatter = ax0.scatter(x=over_50['age'], y=over_50['chol'], c=over_50['target'], cmap='winter')
ax0.set(title='Heart Disease and Cholesterol Levels', ylabel='Cholesterol')
ax0.legend(*scatter.legend_elements(), title='Target')
ax0.axhline(over_50['chol'].mean(), linestyle='--')
ax0.set_xlim([50, 80])

# Second Axes
scatter = ax1.scatter(x=over_50['age'], y=over_50['thalach'], c=over_50['target'], cmap='winter')
ax1.set(title='Heart Disease and Max Heart Rate', xlabel='Age', ylabel='Heart Rate')
ax1.legend(*scatter.legend_elements(), title='Target')
ax1.axhline(over_50['thalach'].mean(), linestyle='--');
ax1.set_ylim([60, 200])

# Figure title
fig.suptitle("Heart Disease Analysis", fontsize=20, fontweight='bold')
```
