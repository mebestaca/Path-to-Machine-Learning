```python
x = [ 1, 2, 3, 4]
y = [10, 30, 20, 40]
plt.plot(x, y)
```

### Different ways to Plot

#### Method 1

```python
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x, y)
plt.show()
```

#### Method 2

```python
fig = plt.figure()
ax = fig.add_axes([0.1, 0.1, 0.8, 0.8])
ax.plot(x, y)
plt.show()
```

#### Method 3

```python
fig, ax = plt.subplots()
ax.plot(x, y);
```

## Sample Matplotlib Workflow

```python
# 0. Import matplotlib and get it ready for plotting
%matplotlib inline
import matplotlib.pyplot as plt

# 1. Prepare Data
x = [ 1, 2, 3, 4]
y = [ 11, 33, 22, 44]

# 2. Set up Plot
fig, ax = plt.subplots(figsize=(8, 8))

# 3. Plot Data
ax.plot(x, y)

# 4. Customize Plot
ax.set(title = "Sample Plot",
      xlabel = "x-axis",
      ylabel = "y-axis")

# 5. Sharing
fig.savefig("sample.png")
```

## Making Figures Using Numpy Arrays

```python
import numpy as np
x = np.linspace(1, 10, 100)
x
```

```python
fig, ax = plt.subplots()
ax.plot(x, x**2);
```

```python
fig, ax = plt.subplots()
ax.scatter(x, np.exp(x));
```

```python
candies_and_prices = { "Apple Candy" : 10, "Banana Candy" : 8, "Lemon Candy" : 15 }
candies_and_prices
```

```python
fig, ax = plt.subplots()
ax.bar(candies_and_prices.keys(), candies_and_prices.values())
ax.set(title="Candy Prices", ylabel="Prices", xlabel="Candies");
```
